## ✅ Comparação entre as classificações dos dois modelos  
### (Multiclasse com binarização vs. Classificação binária direta)

---

### 🔍 Resultados Obtidos:

| Modelo                                      | Accuracy  |
|---------------------------------------------|-----------|
| 🎯 **Rede Multiclasse com Binarização**    | **99,45%** |
| 🛠️ **Rede Binária Direta**                | **92,83%** |

---

🔎 **Conclusão:**  
Ao analisar os resultados, conclui-se que a abordagem de usar uma rede neuronal multiclasse e depois binarizar as predições é mais eficaz do que utilizar uma rede binária direta. A rede multiclasse, após a binarização, obteve uma taxa de acerto de 99,45%, enquanto a rede binária direta alcançou 92,83%. A rede multiclasse, ao ser treinada para distinguir entre 10 classes, aprende representações mais detalhadas das imagens, o que ajuda na classificação binária. Em contraste, a rede binária, ao ser limitada a apenas duas classes, não consegue capturar a complexidade do conjunto de dados de forma tão eficiente. Assim, a rede multiclasse, adaptada para binarizar as predições, oferece um desempenho superior devido à maior capacidade de generalização.
